```python
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers, callbacks
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data
def create_data(samples=1000, features=10):
    X = np.random.randn(samples, features)
    y = np.random.randn(samples, 1)
    return X, y

# Define a neural network model with modified architecture
def create_model():
    model = models.Sequential([
        layers.Dense(64, activation='relu', input_shape=(10,)),  # Increased units for first layer
        layers.Dense(32, activation='relu'),                      # Additional hidden layer with different units
        layers.Dense(16, activation='relu'),                      # Added another hidden layer
        layers.Dense(1)
    ])
    return model

# Training function with Early Stopping and Model Checkpoint
def train_model(model, optimizer, X, y, batch_size, epochs, optimizer_name):
    model.compile(optimizer=optimizer, loss='mean_squared_error')

    # Early stopping to halt training if no improvement in loss
    early_stop = callbacks.EarlyStopping(monitor='loss', patience=10, restore_best_weights=True)

    # Model checkpoint to save the best model during training
    checkpoint = callbacks.ModelCheckpoint(f'{optimizer_name}_best_model.keras', save_best_only=True, monitor='loss')

    # Learning rate schedule for SGD: Reduce learning rate by half every 20 epochs
    if optimizer_name == 'SGD':
        lr_schedule = callbacks.LearningRateScheduler(lambda epoch: 0.01 * 0.5 ** (epoch // 20))
    else:
        lr_schedule = None  # Adam will use its own internal learning rate adjustments

    # Train and record history
    history = model.fit(
        X, y,
        batch_size=batch_size,
        epochs=epochs,
        verbose=0,
        callbacks=[early_stop, checkpoint, lr_schedule] if lr_schedule else [early_stop, checkpoint]
    )

    # Collect loss values and print for comparison
    loss = history.history['loss']
    for epoch, l in enumerate(loss):
        print(f"Epoch {epoch+1}/{epochs} - {optimizer_name} Loss: {l:.4f}")
    return loss

# Generate data
X, y = create_data()

# Create models for SGD and Adam optimizers
model_sgd = create_model()
model_adam = create_model()

# Define optimizers with initial learning rates
optimizer_sgd = optimizers.SGD(learning_rate=0.01)
optimizer_adam = optimizers.Adam(learning_rate=0.001)

# Training parameters
epochs = 100
batch_size = 32

# Train and get loss history for both optimizers
print("\nTraining with SGD Optimizer:")
sgd_loss = train_model(model_sgd, optimizer_sgd, X, y, batch_size, epochs, 'SGD')

print("\nTraining with Adam Optimizer:")
adam_loss = train_model(model_adam, optimizer_adam, X, y, batch_size, epochs, 'Adam')

# Plot loss comparison
plt.figure(figsize=(10, 5))
plt.plot(range(1, len(sgd_loss) + 1), sgd_loss, label='SGD', color='blue')
plt.plot(range(1, len(adam_loss) + 1), adam_loss, label='Adam', color='orange')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('SGD vs Adam Optimizer: Loss Comparison with Early Stopping and Checkpointing')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt

# Load and preprocess the MNIST dataset
(train_imgs, train_labels), (test_imgs, test_labels) = datasets.mnist.load_data()
train_imgs = train_imgs.reshape((train_imgs.shape[0], 28, 28, 1)).astype('float32') / 255
test_imgs = test_imgs.reshape((test_imgs.shape[0], 28, 28, 1)).astype('float32') / 255

# Data Augmentation
datagen = ImageDataGenerator(
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1
)
datagen.fit(train_imgs)

# Build the model
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.3),  # Dropout layer added in dense section
    layers.Dense(10, activation='softmax')
])

# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Set up Early Stopping to stop training if validation loss doesn't improve
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

# Train the model with data augmentation, limited to 5 epochs
hist = model.fit(datagen.flow(train_imgs, train_labels, batch_size=32),
                 epochs=5,  # Reduced number of epochs to 5
                 validation_data=(test_imgs, test_labels),
                 callbacks=[early_stopping])

# Evaluate the model
test_loss, test_acc = model.evaluate(test_imgs, test_labels)
print(f'Test Accuracy: {test_acc}')
print(f'Test Loss: {test_loss}')

# Plot training and validation accuracy
plt.plot(hist.history['accuracy'], label='Train Accuracy')
plt.plot(hist.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Model Accuracy over Epochs')
plt.show()

# Plot training and validation loss
plt.plot(hist.history['loss'], label='Train Loss')
plt.plot(hist.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Model Loss over Epochs')
plt.show()
```
